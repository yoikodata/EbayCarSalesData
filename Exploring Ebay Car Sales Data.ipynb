The above shows the breakdown of cars sold by car brand. For this analysis we are going to take the top 24 car brands and perform our aggregation and mean calculation.

```python
brand_select = autos['brand'].value_counts().nlargest(24).index
print(brand_select)
```

```python
mean_price_dict = {}

for c in brand_select:
    brand = autos[autos['brand'] == c]
    mean_price = brand['price'].mean()
    mean_price_dict[c] = int(mean_price)
    
mean_price_dict
```

```python
mean_mileage_dict = {}

for d in brand_select:
    brand = autos[autos['brand'] == d]
    mean_mileage = brand['odometer_km'].mean()
    mean_mileage_dict[d] = int(mean_mileage)
    
mean_mileage_dict
```

```python
price_series = pd.Series(mean_price_dict)
df = pd.DataFrame(price_series, columns=['mean_price'])

mileage_series = pd.Series(mean_mileage_dict)
df['mean_mileage'] = mileage_series

counts_series = autos['brand'].value_counts()
df['count_sold'] = counts_series

df
```

The top 3 selling cars brands are volkswagen, bmw, and opel (volkswagen by far). The mean price of volkswagen cars sold are $5388 and 128883km respectively. Bmw means price is significantly higher with a slight higher total mileage driven, opel relatively chepaer. The mean_mileage doe not vary as much as price, and some car brands are relatively more expensive despite having higher or similar miles driven. Just looking at the top 6 car brands sold, we see that bmw, mercedes, and audi have a much higher price compared to opel and ford. Volkswagen the highest selling car is priced in the middle.

```python
odo_groups = autos['odometer_km'].value_counts()
print(odo_groups)
```

```python
odo_index = odo_groups[odo_groups > 0].index
odo_index
```

```python
odo_dict = {} 

for o in odo_index:
    odo = autos[autos['odometer_km'] == o]
    mean_price = odo['price'].mean()
    odo_dict[o] = int(mean_price)

#sorting dictionary by key
sorted_odo_dict = {k: v for k, v in sorted(odo_dict.items(), key=lambda item: item[0])}
sorted_odo_dict
```

```python
odo_series = pd.Series(sorted_odo_dict)
df = pd.DataFrame(odo_series, columns=['mean_price'])
df
```

Clear to see price generally decreasing as mileage increases. We see prices are quite small for the low mileage of 5,000. We might have to dig deeper to find out why this is the case. There could also be relatively low data points for this mileage range.

```python
brand_list = []

for x in autos['brand']:
    if x not in brand_list:
        brand_list.append(x)
        
brand_list
```

```python
brand_model_dict = {}

for a in brand_list:
    selected_rows = autos[autos['brand'] == a]
    model_select = selected_rows['model'].value_counts(dropna=False)
    model_select = str(model_select)
    model_select_split = model_select.split()
    model_name = model_select_split[0]
    brand_model_dict[a] = model_name
```

```python
brand_model_series = pd.Series(brand_model_dict)
brand_model_df = pd.DataFrame(brand_model_series, columns=['model_type'])
brand_model_df
```

```python
brand_model_dict_volumes = {}

for a in brand_list:
    selected_rows = autos[autos["brand"] == a]
    model_select = selected_rows["model"].value_counts(dropna=False)
    model_select = str(model_select)
    model_select_split = model_select.split()
    model_volumes = model_select_split[1]
    model_volumes = int(model_volumes)
    brand_model_dict_volumes[a] = model_volumes
    
model_series_vol = pd.Series(brand_model_dict_volumes)
# print(model_series_vol)
brand_model_vol_df = pd.DataFrame(model_series_vol, columns=['model_volumes'])
# brand_model_vol_df
brand_model_df["model_volumes"] = model_series_vol
brand_model_df_vol_sorted = brand_model_df.sort_values("model_volumes", ascending=False)
# brand_model_df_vol_sorted
brand_model_df_vol_sorted
```

We can see which model types above are selling the most

```python
brand_damage_dict = {}

for a in brand_list:
    selected_rows = autos.loc[(autos['brand'] == a) & ((autos['unrepaired_damage'] == 'ja'))]
    damage_mean_price = selected_rows['price'].mean()
    brand_damage_dict[a] = int(damage_mean_price)
    
brand_damage_series = pd.Series(brand_damage_dict)
brand_damage_df = pd.DataFrame(brand_damage_series, columns=['mean_price_damaged'])
brand_damage_df
```

```python
brand_damage_repaired = {}

for a in brand_list:
    selected_rows = autos.loc[(autos['brand'] == a) & (autos['unrepaired_damage'] == 'nein')]
    repaired_mean_price = selected_rows['price'].mean()
    brand_damage_repaired[a] = int(repaired_mean_price)

repaired_series = pd.Series(brand_damage_repaired)
repaired_df = pd.DataFrame(repaired_series, columns = ['mean_price_not_damaged'])

brand_damage_df['mean_price_not_damaged'] = repaired_series
brand_damage_df_sorted = brand_damage_df.sort_values(by='mean_price_not_damaged', ascending=False)
brand_damage_df_sorted
```

```python
brand_damage_df['pct_diff'] = ((brand_damage_df['mean_price_damaged'] / brand_damage_df['mean_price_not_damaged']) * 100).astype(int)
brand_damage_df_final_sorted = brand_damage_df.sort_values(by='pct_diff', ascending = False)
brand_damage_df_final_sorted
```

toyota, dacia, daewoo, mini, lada, skoda are the models that have the largest pecentage of units sold damaged versus undamaged. Clear to see that the mean price of damaged cars are much lower than that that are repaired or undamaged. This is expected.
